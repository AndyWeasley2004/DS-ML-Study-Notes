# Simple Handwritten Number Recognition Model using Multi-dimentional Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
```

```python
# Check dataset
digits = load_digits()

digits.data[0]

plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

### Preprocessing Datasets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, train_size=0.8)
```

### Training

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

### Evaluate Result

Overall Score

```python
model.score(X_test, y_test)
```

Specific Case:

```python
# index 128th element is 9 in the data set
plt.matshow(digits.images[128])
plt.show()

digits.target[128]
```

```python
# use our model to predict
model.predict([digits.data[128]])
# the result is matched.
```

Use confusion matrix to display the detailed performance of the model

```python
y_predict = model.predict(X_test)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict)
cm
```

```python
import seaborn as sns
plt.figure(figsize=[10,7])
sns.heatmap(cm, annot=True)
plt.xlabel('Predict')
plt.ylabel('Actual')
plt.show()
```
